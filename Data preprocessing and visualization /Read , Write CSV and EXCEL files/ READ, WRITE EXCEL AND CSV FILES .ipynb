```python
import pandas as pd
```

```python
df = pd.read_csv("stock_data.csv" , skiprows = 1)
df   #it will skip a row from above and make new row as top
```

```python
df = pd.read_csv("stock_data.csv", header=1, names = ["stock_symbol","eps","revenue","price","people"])
df   #names attribute allows us to write custome column names
```

```python
df = pd.read_csv("stock_data.csv", header=1, nrows=2 )
df   #nrows helps us to specify how many rows we want to show               
```

```python
df = pd.read_csv("stock_data.csv",  na_values={
        'eps': ['not available'],
        'revenue': [-1],
        'people': ['not available','n.a.']
    }, header=1)
df  #here we have to specify which values we want to be treated as na values
```

```python
df = pd.read_csv("stock_data.csv", na_values=["n.a.", "not available"], header=1)
df
```

```python
df["pe"] = df["price"] / df["eps"]
df   #here we have derived a new column called pe 
```

```python
df.to_csv("pe1.csv" , index=False , header=False)  #it will create a new csv file called pe1.csv and export all the data over there with a new column also included called pe
```
